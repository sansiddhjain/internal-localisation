```python
from ast import literal_eval
from random import randint
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap
import pandas as pd
import utility
```

```python
# Reading Details of access points in to a dictionary
APs = pd.read_csv("location_APs_feets.csv")
APs = APs.set_index('Controller Id').T.to_dict('list')
## Now Aps is a dict with AP number as key with value a list [ floor, X, Y]
```

```python
def plot(circle):
    color = ['b','g','c']
    fig, ax = plt.subplots()
    plt.xlim((0,110))
    plt.ylim((0,70))
    for i,c in enumerate(circle):
        temp = plt.Circle((c[0], c[1]), c[2], alpha=0.5, color=color[i])
        ax.add_artist(temp)
    plt.show()
```

```python
file = "best_mac_mod_logs/2018-1-15/log.csv"
df = pd.read_csv(file, converters={'controllerid':literal_eval, 'pwr':literal_eval, 'ts':pd.to_datetime})
print df.dtypes
for i in df.index:
    now = df.loc[i]
    mac = now['mac']
    controllerids = now['controllerid']
    powers = now['pwr']
    
    probes_per_floor = {}
#     print powers
    # seperate controller into floors
    for (controller, power) in zip(controllerids, powers):
        floor = int(APs[controller][0])

        if floor not in probes_per_floor:
            probes_per_floor[floor] = []
        
        distance_from_controller = utility.signal_strength_to_distance(float(power), 2.4)
        
        # Doubt: remove distances greater than 130(max distance per floor) - could indicate signals from different floor or from outside
        if distance_from_controller < 130:
            probes_per_floor[floor].append(APs[controller][1:]+[distance_from_controller])
    
#     print probes_per_floor, "\n"

    # most probable floor: the one with max probes
    most_probable_floor = None
    temp = 0
    for key in probes_per_floor.keys():
        if len(probes_per_floor[key]) > temp:
            temp = len(probes_per_floor)
            most_probable_floor = key
    if most_probable_floor is not None:
        most_probable_floor = probes_per_floor[most_probable_floor]
        if len(most_probable_floor) >= 2:
#             print most_probable_floor
            plot(most_probable_floor)
        
    
```
